### 1. 선형회귀

선형회귀 분석은 독립변수(independent: x)와 종속변수(dependent: y) 간의 관계를 찾는 통계적 방법이다.  
또는 y와 한개이상의 x의 선형상관관계를 모델링하는것이다.<br>

가장 간단한 회귀식
$$Y = AX + B$$

여기서 X, Y는 데이터이고 A, B는 두 데이터 사이의 관계를 설명하는 Paramenter가 된다.  
이러한 관계식을 통계에 적용하는 행위를 **회귀**라고 부른다.<br>
<br>
그러나 통계정보는 이런 분명한 관계를 얻을 수 없다.  
그래서 아래와 같이 수정하여 사용한다.  
<br>
$$Y = AX + B + e$$

위 항의 e는 Error라고 하며, 컴퓨터로 회귀계산을 실행하면 자동적으로 e의 평균이 0이 되는 조건을 만족하는 A와 B를 추정하게 된다.  <br>
<br>
즉, 회귀식이란 e의 평균이 0이 되는 조건을 만족하는 A와 B의 관계식인 것이다.(e의 평균이 최소가 되도록 parameter의 값이 점차 좁혀지기 때문에 회귀라고 부른다)

회귀 연산을 통해 두변수 X와 Y사이의 관계를 추정이 가능하다.  
ex) x가 증가하면 y도 증가.

### 2. R-suqare
회귀만을 이용해서는 x값으로 y 값을 알아 낼수 없다.  
다만 빨간 직선이 y가 될 수 있는 평균 추정치를 제공할 뿐이다.<br>
<br>
이런 단서를 통해서 회귀식의 정확도를 테스트가 가능하다.<br>
<br>
$$y = (Ax+b)+(e)$$

로 항을 2개로 나눴을 때. $(Ax + b)$는 회귀로 얻게되는 정보이고 $(e)는 알 수 없는 정보이다.  
만약 앞의 항이 뒤의 항보다 강하게 작용하면 식은 정보를 분석하는데 매우 좋은 모델이라 말할 수 있다.  
반대로 뒤의 항이 앞의 항보다 강하게 작용할 경우에는 매우 부정확한 모델이라 할 수 있다.  <br>
<br>
위의 정도를 수치로 표시한 것을 R^2, 또는 R-Square라고 부른다.  
R-Square가 1이면 e의 값이 0인 존재하지 않은 상태이고,  
R-Square가 0이면 e의 값이 1인 $Ax + b$가 무의미한 상태를 말한다.  <br>
<br>

## 시계열 분석-개요

시계열은 과거, 현재에 대한 분석에 매우 정확하다.  
그러나 미래에 대해서는 부정확 할 확률이 높으니 조심해야한다.  
특히 가까운 미래보다 먼미래에 대해서 더 부정적이다.  
그러나, 미래에 대해서 알 수는 없지만, 미래에 대한 시나리오를 비교하는데는 사용이 가능하다.  
또한 시계열로 대상의 특징이나 다른 요소와의 관계를 규명할 수 있다.  

## 시계열 분석-모형

시계열 분석과 회귀분석의 가장 큰 차이점은 시간의 고려 여부이다.  
시계열 정보는 크게 2가지로 나뉘는데
* 규칙성을 가지는 패턴
    - 이전의 결과와 이후의 결과 사이에서 발생하는 자기상관성(Autocorrelativeness)
    - 이전에 생긴 불규칙한 사건이 이후의 결과에 편향성을 초래하는 이동평균(Moving Average)현상.
* 불규칙성을 가지는 패턴  
  
으로 분리된다.  <br>
<br>
한편 많은 수계열 모형들이 불규칙한 패턴을 White Noise라고 칭하고 평균이 0이며 일정한 분산을 지닌 정규분표에서 추출된 임의의 수치라고 규정하고 있는데,  
이러한 정규분포가정은 모델의 해석을 전반적으로 편리하게 만들기 때문에 대부분의 통계 및 수리 분석에서 채택되고 있다.

### 1. 이동평균(Moving Average)-MA모형

시간이 지날수록 어떤 Random Varianble의 평균값이 지속적으로 증가하거나 감소하는 경향이 생길 수 있다.  
예를 들어   
봄에서 여름이 될 수록 가계 전기 수요량은 증가하는 경향이 있고,  
여름에서 겨울로 갈 수록 감소하는 경향이 있다.  
특별히 전기세가 많이 부과 되지 않았다면 전월의 전기 사용량이 다음월 전기 사용량에 상관을 주지는 않을 것이라고 가정할 수 있다.  
하지만 전기 사용량이 얼마가 될 것인지는 확실하게 측정하기 어렵기 때문에 이러한 경우 평균이동이 있는 시계열 데이터가 될 것이라고 생각해볼수 있다.  
ex)
맥주 소비량에 따른 평균값의 변화

![](https://seongjaemoon.github.io/assets/uploads/algorithm/maGragh2.jpg)

데이터의 평균값이 시간에 따라 변화하는 경향이 Moving Average이다.

이동평균을 시계열 모형으로 구성한 것을 MA모형이라고 부르는데,  
가장 간단한 형태가 바로 직전 데이터가 다음 데이터에 영향을 준다고 가정한 MA(1)모형이다.  
$$
X(t) = (a \times e(t-1) + c) + u \times e(t)
$$


시점 $t$에서 얻게 될 $X(t)$의 평균값은 
시점 $t-1$에서 발생한 $errore(t-1)$의 값에 $a$를 곱하고 $c$를 더한 것과 같다는 뜻이다.  
즉, $X(t)$값은 평균이 $a \times e(t-1)$이며 분산이 $u$인 정규 분포에서 도출되는 임의의 값이라는 뜻이다.  
뒤에 나올 AR모형과의 가장 큰차이는 이전에 발생한 error가 중요한 것이지 결과적으로 X(t-1)가 무엇이였는지는 중요하지 않다는 점이다.

  [그림이미지 6](https://www.investopedia.com/terms/s/sma.asp)  
최근 50일 평균값보다 최근 15일 이동 평균값이 커지면 주가가 치솟는다.  
즉, 골든크로스가 발생한다.

### 2. 자기상관(Autocorrelation)-AR모형
자기 상관이란 어떠한 Random Variable에 대해서 이전의 값이 이후 값에 영향을 미치고 있는 상황을 이야기한다.  
ex)  
이전에 값이 상승 이후에 값 하락<br>
<br>
하는 경향을 말한다.  
자기상관은 바로 이전 결과의 영향을 받기도 하지만 드물게 Delay가 발생하기도 한다.  
우리가 실생활에서 가장 많이 발견하는 평균으로 돌아가려는 경향을 지닌 것들을 말한다.  
<br>
ex)고무줄, 용수철 등등.  
잡아 당기면 다시 원래상태로 복귀하려는 움직임을 보이는 상태를 말한다.  
<br>
이러한 자기상관성을 시계열 모형으로 구성한 것이 AR모형이라고 부르는데, 가장 간단한 형태가 바로 직 전데이터가 다음 데이터에 영향을 준다고 가정한 AR(1)모형이다.  
$$
X(t) = {a \times X(t-1) + c} + u \times e(t)
$$


식에 대해 설명하면 
시점 t에서 얻게 될 $X(t)$의 평균값은   
시점 t-1에서 얻었던 $X(t-1)$의 값에 $a$를 곱하고 $c$를 더한것과 같다는 말이다.  
여기서 $e(t)$항은 white noise라고 부르며 평균이 0이고 분산이 1인 정규분포에서 도출된 random한 값이다.  
즉, $X(t)$ 값은 평균이 $a \times X(t-1)+c$이며 분산이 $u$인 정규 분포에서 도출되는 임의의 값이라는 뜻이다.  
<br>
만약 더 이전의 시점(P시점)을 모델에 넣고자 하면 $AR(P)$모형이 되는 것이다.  
$$
X_t = c + \sum_{i=1}^p \varphi_i X_{t-1} + \varepsilon_t^*
$$

### 3. ARMA모형
ARMA모형은 Autoregressive Moving Average라는 뜻으로,  
AR모형과 MA모형을 합친 것으로,  
연구 기관에서 가장 선호되는 모델이다.(통상 ARMA(2, 2) 정도의 모델이면 충분히 양질의 데이터를 얻는다)  
<br>
가장 단순한 형태인 ARMA(1, 1)모형은 아래와 같다.  
  
$$
    X(t) = (a \times X_{t-1}) + (b \times e_{t-1}) + c + u \times e(t)
$$
  
ARMA(2, 2)모형은 아래와 같다
  
$$
X(t) = (a_1 \times X_{t-1} + a_2 \times X_{t-2}) + (b_1 \times e_{t-1} + b_2 \times e_{t-2}) + c + u \times e(t)
$$
  
이후 ARMA(p, q)는 예상과 같이 점점 $a_i, x_{t-i}, e_{t-i}$ 변수가 늘어난다.

### 4. ARIMA모형
ARIMA(Autoregressive Integrated Moving Average)라는 뜻으로,  
ARMA모형이 과거의 데이터들만 사용하는 것에 반해 과거의 데이터가 가지고 있는 '추세(Momentum)'까지 반영한 모델이라 할 수 있다.  
즉, Correlation 뿐 아니라 Cointegration까지 고려한 모델이다.  
  
> **선형관계**    
    두 변수 X-Y간에 correlation이 0보다 크면 X가 큰 값이 나올때 Y값도 큰 값을 가진다.  
    두 변수 X-Y간에 correlation이 0보다 작으면 X가 큰 값이 나올때 Y값은 작은 값을 가진다.  <br>

---
> **추세관계**  
    두 변수 X-Y간에 cointegration이 0보다 크면 X의 값이 이전 값 보다 증가하면 Y값도 증가한다.  
    두 변수 X-Y간에 cointegration이 0보다 작으면 X의 값이 이전 값 보다 증가하면 Y값은 감소한다.

Correlation이 0보다 작고 cointegration이 0보다 큰 관계일때,  
X가 큰값이면서 증가하는 추세에 있는 경우  
Y는 현재 작은 값이나 빠르게 증가하는 추세로 반응하게 된다.  <br>
<br>
Correlation이 0보다 크고 cointegration이 0보다 작은 관계일때,  
X가 큰 값이면서 증가하는 추세에 있는 경우  
Y는 현재 큰 값이나 빠르게 감소하는 추세로 반응하게 된다.  

<br>
ARIMA 모형은 추세를 고려할 수 있기 때문에 momentum을 중요하게 보는 분석가들에게 매우 유용하다.  
  
```
ARIMA 모델은 자기 자신의 추세만 고려한다.  
white noise의 추세는 고려하지 않는다.  
올바른 모델의 white noise는 추세가 존재하면 안되기 때문이다.  
즉, Autoregressive는 자기 자신의 correlation을 말하는 것이고, 
Integrated 모델은 자기 자신의 cointegration을 말하는 것이다. 
```

ARIMA모델 식 ARIMA(1, 1, 1)
$$
a \times (X_{t} - X_{t-1}) = (b \times X_{t-1}) + (c \times e_{t-1})+ d + u \times e_t
$$

즉, X의 추세는 이전 X의 값과 이전의 white noise의 결과의 영향을 받게 된다는 뜻이다.  
식을 조금 다르게 전개하면 아래와 같은 형태가 된다.  
<br>
$$
X(t) = \frac{X_{t-1} + (b \times X_{t-1}) + (c \times e_{t-1}) + d + u \times e_t}{a}
$$

사실상 ARIMA(1, 1, 1)은 ARMA(1, 1)와 거의 비슷하여 사용이 매우 적다.  
따라서 ARIMA(1, 2, 1)를 많이 사용한다.  
  
$$
a \times ((X_t - X_{t-1}) - (X_{t-1} - X_{t-2})) = (b \times X_{t-1}) + (c \times e_{t-1}) + d + u \times e_t
$$

위의 모델은 X를 2차 미분한 값에 대한 모델이다.  
데이터가 어떤 모멘텀 성향을 지닌다고 가정했을 때 모멘텀의 변화를 모델링한 것이라고 보는 것이 맞다.  
즉, 이 모델의 모멘텀 변화는 X의 이전 값과 이전에 발생한 white noise에 의해 결정된다는 것을 표현 한것이다.  
<br>
마찬가지로 $X_t$에 대해서만 전개하면 아래와 같은 수식이 되며, 이는 ARMA(2, 1)과 매우 유사하다.  
<br>
$$
X_t = (2 + \frac{b}{a}) \times X_{t-1} + X_{t-2} + \frac{c}{a} \times e_{t-1} + \frac{d}{a} + \frac{u}{a} \times e_t
$$

ARIMA는 수식의 복함함때문에 많이 사용되지 않으며 추세가 유의미 하지 않을 경우에는 ARMA모형을 택하는 경우가 많다.

````python
# Define the p, d and q parameters to take any value between 0 and 2
p = d = q = range(0, 2)

# Generate all different combinations of p, q and q triplets
pdq = list(itertools.product(p, d, q))

# Generate all different combinations of seasonal p, q and q triplets
seasonal_pdq = [(x[0], x[1], x[2], 12) for x in list(itertools.product(p, d, q))]

print('Examples of parameter combinations for Seasonal ARIMA...')
print('SARIMAX: {} x {}'.format(pdq[1], seasonal_pdq[1]))
print('SARIMAX: {} x {}'.format(pdq[1], seasonal_pdq[2]))
print('SARIMAX: {} x {}'.format(pdq[2], seasonal_pdq[3]))
print('SARIMAX: {} x {}'.format(pdq[2], seasonal_pdq[4]))
````

````python
warnings.filterwarnings("ignore") # specify to ignore warning messages

for param in pdq:
    for param_seasonal in seasonal_pdq:
        try:
            mod = sm.tsa.statespace.SARIMAX(y,
                                            order=param,
                                            seasonal_order=param_seasonal,
                                            enforce_stationarity=False,
                                            enforce_invertibility=False)

            results = mod.fit()

            print('ARIMA{}x{}12 - AIC:{}'.format(param, param_seasonal, results.aic))
        except:
            continue
````

````python
mod = sm.tsa.statespace.SARIMAX(y,
                                order=(1, 1, 1),
                                seasonal_order=(1, 1, 1, 12),
                                enforce_stationarity=False,
                                enforce_invertibility=False)

results = mod.fit()

print(results.summary().tables[1])
````
